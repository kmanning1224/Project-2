```python
import pandas as pd
from pymongo import MongoClient
```

```python
client = MongoClient("mongodb://localhost:27017/")
```

```python
#database
db = client["assault_db_us"]
```

```python
#collection
collection = db["assault_us"]
```

```python
data = pd.read_csv("new_global_dataset_2020.csv")
```

```python
data.reset_index(inplace=True)
data_dict = data.to_dict("records")
collection.insert_one({"index":"ID", "data":data_dict})
```
